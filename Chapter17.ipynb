## Rayleigh-Ritz Method

The Rayleigh-Ritz method is based on an elegant branch of mathematics, the calculus of variations. In this method we solve a boundary-value problem by approximating the solution with a finite linear combination of simple basis functions that are chosen to fulfill certain criterion, including meeting the boundary conditions.

The calculus of variations seeks to optimize (often minimize) a special class of functions called *functionals*. A functional can be thought of as a "function of functions." The usual form for the functional is 

$$
I[y] = \int_a^b F(x,y,\frac{dx}{dy})dx
$$

The bracket [ ] notation is utilized to highlight that $I[y]$ is not a variable, but a function $y=y(x)$. The value of the right-hand side of the equation above will not change as the function $y(x)$ is varied, but when $y(x)$ is fixed, it evaluates to a scalar quantity (a constant).

Let's illustrate this concept with a simple example where the solution is obvious. We will find the funciton $y(x)$ that minimizes the distance between two points. We know the answer is a straight line, but we will pretend we don't. In this case the functional is the integral of the distance along any of these curves:

$$
I[y] = \int_{x_1}^{x_2}\sqrt{(dx)^2 + (dy)^2}dx = \int_{x_1}^{x_2}\sqrt{1 + (\frac{dy}{dx})^2}dx
$$

To minimize $I[y]$ we set it's derivative to zero. There are certain restrictions on the possible solution. Obviously each must pass through the points $(x_1, y_1)$ and $(x_2, y_2)$. In addition, for the optimal trajectory, the Euler-Lagrance equation must be satisfied:

$$
\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right) - \frac{\partial F}{\partial y} = 0
$$

Applying this to the functional for shortest distance, we have

$$
F(x,y,y') = \sqrt{1+(y')^2} \\
\frac{\partial F}{\partial y} = 0 \\
\frac{\partial F}{\partial y'} = \frac{1}{2} \left(1 + (y')^2\right)^{-\frac{1}{2}}2y'\\
\frac{d}{dx} \frac{\partial F}{\partial y'} = \frac{d}{dx} \left( \frac{y'}{\sqrt{1+(y')^2}}\right) = 0
$$

From this it follows that, whatever is inside the parentheses in the last equation above, must be equal to a constant.

$$
\frac{y'}{\sqrt{1+(y')^2}} = c
$$

Or if solved for $y'$, the expression is again some constant, say $b$

$$
y' = \sqrt{\frac{c^2}{1-c^2}} = b 
$$

Or after integration, 
$$
y = bx + a
$$

We can use the boundary conditions to evaluate the constants $a$ and $b$.

We will now advance to a less trivial case. Consider the second-order linear boundary-value problem over $[a, b]$. 

$$
y'' + Q(x)y = F(x), \hspace{5pt} y(a) = 0, \hspace{5pt} y(b) = 0 \hspace{5pt} (*)
$$

It turns out that the functional that corresponds to $(*)$ is

$$
I[u] = \int_a^b\left[\left(\frac{du}{dx}\right)^2 - Qu^2 + 2Fu\right] dx \hspace{5pt} (**)
$$

While this is presented without proof we can verify this is true by transforming the integrand of $(**)$ into $(*)$ through the Euler-Lagrange equation. Observe the benefit of using $(**)$ rather than $(*)$, we now only have first-order derivatives instead of second order-ones. This not only simplifies the mathematics but permits finding solutions even when there are discontinuities that cause $y$ to not have sufficiently high-derivatives. 

If we know the solution to our differential equations, substituting it for $u$ in $(**)$ will make $I[u]$ a minimum. When the answer isn't known, we can approximate it by some *admissible* function (meaning it complies with certain boundary conditions). The Rayleigh-Ritz method is based ont his idea. We assume that $u$ can be approximated by a lienar sum of functions:

$$
u = c_1 v_1 + c_2 v_2 + ... + c_n v_n = \sum_{i=1}^n c_i v_i
$$

We substitute $u$ and $\frac{du}{dx}$ from this into $(**)$ to get

$$
I(c_1, c_2, ..., c_n) = \int_a^b \left[ \left(\frac{d}{dx} \sum c_i v_i\right)^2 - Q\left(\sum c_iv_i\right)^2 + 2F\sum c_i v_i\right] dx
$$

At this point, observe that $I$ is an ordinary function of the unkown $c_i$'s. To get the optimum, we differentiate with respect to each $c_i$ in turn, setting each to zero. This gives, when the integrals are evaluated, a linear system in the $c_i$'s that we solve by the methods we have already learned.

## An Example

Let's use the example we looked at in the last section of a simply supported beam of length $L$ with a load $P$ acting at the location $\frac{L}{2}$. The boundary value problem is written as:

$$
El y''(x) = -\frac{Px}{2}, \hspace{5pt} y(0)=0, \hspace{5pt} y(L) = 0
$$

Which we have already shown has the following solution (here shown from *Python*):

Now we will use the Rayleigh-Ritz method to show we can achieve the same solution as the analytic one through the proper choice of an approximating function. First we define the term $F$ from equation $(**)$ as follows:

$$
F = -P x / 2 / EI ;
$$

For an approximation to the deformation we will use a cubic polynomial. (It turns out that the smallest polynomial we can use and recover the exact solution is cubic. This is because it satisfies both the essential and natural boundary conditions.)

$$
u[x] = c_1 + c_2 x + c_3x^2 + c_4x^3
$$

Applying the boundary conditions and solving for the constants in terms of one another, we now have an admissible funciton:

$$
c_2 \rightarrow -c_3L - c_4L^2, \hspace{5pt} c_1 \rightarrow 0 \\
(-c_3L -c_4L^2)x + c_3x^2 + c_4x^3
$$

Now we can find the constants which make $I[u]$ stationary.

$$
\frac{2c_3L^3}{3} + c_4 L^4 + \frac{L^4 P}{12EI}\\
c_3L^4 + \frac{8c_4L^5}{5} + \frac{2L^5 P}{15EI}
$$

The results in two equations and two unkowns which we can solve for directly.

$$
c_3 \rightarrow 0, c_4 \rightarrow - \frac{P}{12EI}
$$

Substituting these values of the constants back into our assumed dispacement function, we can see that we recover the exact analytic solution.

$$
\frac{Px\left(L^2-x^2\right)}{12EI}
$$

As an aside, it turns out that we can also use for our functional $I[u]$ the total potential energy of the system we are solving. We can minimize this functional. For this reason, particular cases of the method described above are sometimes called the *Principle of Stationary Potential Energy.*

## Finite Element form of Rayleigh-Ritz Method

The problem with the classical form of the Rayleigh-Ritz method from variational principles is that the proper form of approximating function is not always easy to find. Generally, only polynomials and occasionally sine and cosine functions are simple enough to be practicable, but these functions do not always approximate the dependent variable correctly. The trouble lies in trying to fit a complex function over the entire interval $[a,b]$.

We have previously seen taht even low order polynomials can approximate any funciton if applied over a small enough region. For the FE form of the Rayleigh-Ritz method we do something equivalent to interpolations between values of the solution at distinct points within $[a,b]$. If we knew values for $y$ at these points, we could interpolate to get $y$ at points between; if we assume a linear relationship we can get $y$ int he subinterval $[x_i, x_{i+1}]$ from $y_i$ and $y_{i+1}$ by defining:

$$
y(x) = N_i + N_{i+1}y_{i+1}, \text{ with}\\
N_i = \left\{ \begin{matrix} \frac{(x-x_{i-1})}{h_{i-1}} & \text{on }[x_{i-1}, x_{i}] \\ \frac{(x_{i+1}-x}{h_{i}} & \text{on }[x_{i}, x_{i+1}] \\ 0 & \text{elsewhere} \end{matrix}\right\}
$$

where $h_i = |x_{i+1} - x|$. In effect, $y$ within the interval is a weighted average of the values at the ends. Here are a few properties of these weighting functions:

  1. $N_i(x_j) = \left\{\begin{matrix}1 & \text{if } i = j \\ 0 & \text{otherwise} \end{matrix}\right\}$
  1. $N_i(x)*N_j(x) = 0$ for $|i-j|>1$ (intervals other than those that have $x_i$ as an endpoint)
  1. $N_i'(x)*N_j'(x) = 0$ for $|i-j|>1$
  1. $N_i(x)*N_{i+1}(x) = \left\{\begin{matrix} \frac{(x_{i+1}-x)(x-x_i)}{h^2} & \text{on }[x_i,x_{i+1}]\\ 0 & \text{elsewhere} \end{matrix}\right\}$
  1. $N_i'(x)*N_{i+1}'(x) = \left\{\begin{matrix} \frac{-1}{h^2} & \text{on }[x_i, x_{i+1}] \\ 0 & \text{elsewhere} \end{matrix}\right\}$
  
where $h$ is a constant in the relationships above.

We will approximate the solution to the differential equation by substituting into the function for $u(x)$:

$$
u(x) = \sum_{i=1}^nc_iN_i(x)
$$

When we do so and minimize by setting derivatives with respect to $c_i$ equal to zero, the values of the $c_i$'s will be approximations to $y$ at the nodes within $[a,b]$

## FE Form of Rayleigh-Ritz Example

Solve $y''-\left(1-\frac{x}{5}\right)y = x, \hspace{5pt}y(1)=2, \hspace{5pt}y(3)=-1$

We will subdivide the interval $[1,3]$ into four equal subintervals, making $h=0.5$. We approximate the function with:
$$
y(x) = c_0N_0(x)+c_1N_1(x)+c_2N_2(x)+c_3N_3(x)+c_4N_4(x)
$$
$c_0$ will be $2$ and $c_4$ will be $-1$ to satisfy the boundary conditions, but we will postpone this substitution. We now substitute the approximation into the functional as before

$$
I(c_0, ..., c_4) = \int_1^3\left[(c_0N_0'(x)+c_1N_1'(x)+c_2N_2'(x)+c_3N_3'(x)+c_4N_4'(x))^2+\left(1-\frac{x}{5}\right)(c_0N_0(x)+c_1N_1(x)+c_2N_2(x)+c_3N_3(x)+c_4N_4(x))^2+2x(c_0N_0(x)+c_1N_1(x)+c_2N_2(x)+c_3N_3(x)+c_4N_4(x))\right]dx
$$

Now we set the derivatives of $I$ with respect to $c_i, i = 1,2,3$ to zero. We do not do this for $c_0$ and $c_4$ because they are fixed. We also substitute the known values for $c_0$ and $c_4$.

This gives,
$$
A\bar{c} = \bar{b}
$$
where
$$
A_{ik} = \int_1^3\left[N_{k}'(x)N_i'(x) + \left(1-\frac{x}{5}\right)N_k(x)N_i(x)\right]dx\\
b_k = -\int_1^3 xN_k(x)dx
$$
we won't work out all terms, but a few are provided for verification by the student.

$$
A_{11} = 4 + 4\int_{1.0}^{1.5}\left(1-\frac{x}{5}\right)(x-1)^2dx + 4\int_{1.5}^{2.0}(2-x)^2dx\\
A_{12} =-2 + 4\int_{1.5}^{2.0}\left(1-\frac{x}{5}\right)(2-x)(x-1.5)dx \\
b_1 = -2 \int_{1.0}^{1.5}x(x-1)dx - 2 \int_{1.5}^{2.0}x(2-x)dx
$$

If we use Simpson's method to evaluate the integrals, we get
$$
A = \begin{pmatrix}4.233 & -1.9458 & 0.0\\-1.9458 & 4.1999 & -1.9542 \\0.0 & -1.9541 & 4.1667 \end{pmatrix}, \hspace{5pt} b=\begin{pmatrix}3.125 \\ -1.0 \\ -3.2125 \end{pmatrix}
$$

we can solve using Gaussian elimination to get,
$$
c=\begin{pmatrix}-0.5323\\-0.4480\\-0.9811\end{pmatrix}
$$
Now we have an approximation to the solution of our differential equation,
$$
y(x) = 2N_0(x)-0.5323N_1(x)-0.4480N_2(x)-0.9811N_3(x)-N_4(x)
$$
The nodal values are obious,
$$
y(1.5) = 0.5323\\
y(2.0) =-0.4480\\
y(2.5) =-0.9811
$$
but we can also use the equation for $y$ to compute for any point in $[1,3]$.
