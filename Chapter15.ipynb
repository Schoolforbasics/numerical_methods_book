## Systems of Ordinary Differential Equations

Many higher order differential equations can be transformed into a system of first order differential equations by allowing the higher order derivatives to be new functions. This is best illustrated with an example. Consider the equation:

$$
A \frac{d^3y}{dt^2} + B \frac{d^2y}{dt^2} + C \frac{dy}{dt} + D y = 0
$$

If we let $z = \frac{d^2y}{dt^2}$ we have:

$$
\frac{dz}{dt} = \frac{1}{A} \left(-Bz - C \frac{dy}{dt} - Dy\right)
$$

and letting $x = \frac{dy}{dt}$, we have the following system of first order differential equations:

$$
\frac{dx}{dt} = z\\
\frac{dy}{dt} = x\\
\frac{dz}{dt} = \frac{1}{A}(-Bz - Cx - Dy)
$$

We can now simply use generalizations of techniques we have already learned for solving a single first order equation to solve this system of first order equations. The procedure foollows exactly as before except now we alternate (for the example above) through the $x, y,$ and $z$ calculations.

## Pseudocode for Runge-Kutta Method for Systems of Differential Equations

To approximate the solution of the $m^{th}$-order system of first-order differential equations with initial values
$$
u'_j = f_j (t, u_1, u_2, ..., u_m), \hspace{5pt} a\leq t \leq b, \hspace{5pt} with \hspace{5pt} u_j(a) = \alpha_j
$$
for $j = 1, 2, ..., m$ at $(N+1)$ equally spaced numbers in the interval $[a,b]:$

  1. Set $h = \frac{b-a}{N}$
  1. Set $t = a$
  1. For $i = 1, 2, ..., N$ do Steps A-F
    1. For $j = 1, 2, ..., m$ do Step a
      1. Set $k_{1, j} = hf_j (t, w_1, w_2, ..., w_m)$
    1. For $j = 1, 2, ..., m$ do Step a
      1. Set $k_{2,j} = hf_j \left(t + \frac{h}{2}, w_1 + \frac{1}{2}k_{1,1}, w_2 + \frac{1}{2}k_{1,2},..., w_m + \frac{1}{2}k_{1,m}\right)$
    1. For $j = 1, 2, ..., m$ do Step a
      1. Set $k_{3,j} = hf_j \left(t + \frac{h}{2}, w_1 + \frac{1}{2} k_{2,1}, w_2 +\frac{1}{2} k_{2,2}, ..., w_m + \frac{1}{2}k_{2,m} \right)$
    1. For $j = 1, 2, ..., m$ do Step a
      1. Set $k_{4,j} = hf_j \left(t + h, w_1 + k_{3,1}, w_2 + k_{3,2}, ..., w_m + k_{3,m} \right)$
    1. For $j = 1, 2, ..., m$ do Step a
      1. Set $w_j = w_j + \frac{k_{1,j} + 2k_{2,j} + 2k_{3,j} + k_4}{6}$
    1. Set $t = a + ih$

## Remarks about *Stiff Equations*

A *stiff equation* results from phenomena with widely differing time scales. For example, the general solution of a differential equation may involve sums or differences of terms of the form $a e^{ct}, be^{dt},$ where both $c$ and $d$ are negative (and therefore convergent) but $c$ is much smaller than $d$. In such cases, using a small value for the step size can introduce enough round-off error to cause instability. An example is the following:

$$
\frac{dx}{dt} = 1195 x - 1995y, \hspace{5pt} x(0) = 2 \\
\frac{dy}{dt} = 1197 x - 1997y, \hspace{5pt} y(0) = -2
$$

The analytic solution to this system of equations is:

$$
x(t) = 10e^{-2t} - 8e^{-800t}\\
y(t) = 6e^{-2t} - 8e^{-800t}
$$

Now let's suppose we wanted to attempt to solve this system of equations with Euler's method, we would have:

$$
x_{i+1} = x_i + hf(x_i, y_i) = x_i + 0.1(1197x_i - 1995y_i)\\
y_{i+1} = y_i + hf(x_i, y_i) = x_i + 0.1(1197x_i - 1997y_i)
$$

This gives $x(0.1) = 640, y(0.1) = 636,$ while the exact values are $x(0.1) = 8.187$ and $y(0.1) = 4.912$. Results like this are typical for stiff equations.

One solution to this problem is to use an *implicit* method rather than an explicit one. All of the methods so far discussed have been explicit, meaning that the new values, $x_{i+1}$ and $y_{i+1}$, are computed in terms of previous ones, $x_i$ and $y_i$. The implicit form of Euler's method is

$$
x_{i+1} = x_{i} + hf(x_{i+1}, y_{i+1})\\
y_{i+1} = y_{i} + hf(x_{i+1}, y_{i+1})
$$

If the derivative functions $f(x,y)$ and $g(x,y)$ are nonlinear, this will require an iterative solution to find the roots, but in our example above, they are linear and we have:

$$
x_{i+1} = x_i + 0.1(1195x_{i+1} - 1995y_{i+1})\\
y_{i+1} = y_i + 0.1(1197x_{i+1} - 1997y_{i+1})
$$

or

$$
\begin{pmatrix} x_{i+1} \\ y_{i=1} \end{pmatrix} = \begin{pmatrix} (1-1195(0.1)) & 1995(0.1)\\-1197 & (1+1997(0.1)) \end{pmatrix}^{-1} \begin{pmatrix}x_i \\ y_i \end{pmatrix}
$$

which has the solution $x(0.1) = 8.23, y(0) = 4.90$, reasonably close to the analytical values, of course, calculating the inverse will take additional computational effort at every step.

## Explicit vs. Implicit and Stability

In the previous slide we saw both the explicit and implicit forms of the simple Euler method. Let's investigate the stability of these methods in the context of an example. Consider the equation:

$$
\frac{dx}{dt} = \lambda x, \hspace{5pt} x(0) = x_0
$$

which of course has the analytic solution

$$
x(t) = x_0 e^{\lambda t}
$$

We also know that this solution is only bounded (stable) when $R(\lambda) \leq 0$. Let's keep this in mind as we investigate the stability of the numerical scheme. First, Euler explicit method:

$$
\begin{matrix} x_{n+1} & = & \lambda h x_n + x_n \\ & = & (1+\lambda h)x_n \\ & = &(1+ \lambda h)^2 x_{n-1} \\ & \vdots & \\ & = & (1+\lambda h)^{n+1} x_0 \end{matrix}
$$

This result shows us that as $n$ increases the only way for $x_{n+1}$ not to grow in an unbounded fasion is for the following to be true:

$$
\left|1+\lambda h\right| \leq 1
$$

We can solve this equation for $h$ to give us a critical step size in order to ensure stability

$$
h \leq \left| \frac{2}{\lambda} \right|
$$

Now lets do the same thing for the implicit scheme.

$$
\begin{matrix} x_{n+1} & = & \lambda h x_{n+1} + x_n \\ x_{n+1} - \lambda h x_{n+1} & = & x_n \\ (1-\lambda h) x_{n+1} & = & x_n \\ x_{n+1} & = & \frac{x_n}{1-\lambda h} \\ & = & \frac{x_{n-1}}{(1-\lambda h)^2} \\ & \vdots & \\ & = & \frac{x_0}{(1-\lambda h)^{n+1}}\end{matrix}
$$

This result is stable as long as the following inequality holds true.

$$
\left|1-\lambda h \right| \leq 1
$$

Because of the retriction on $\lambda$ that requres the $R(\lambda) \leq 0$, the inequality will always be true. This is said to be unconditionally stable. So in summary, explicit schemes require less computational effort at each step, but require a small step size for stability. Implicit methods require more computational effort at each step, but can utilize much larger step sizes and under certain conditions can be unconditionally stable.
